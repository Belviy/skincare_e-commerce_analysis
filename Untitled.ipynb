```python
#1.	What were the most profitable segments from 2020 to 2023
#Step 1: Filter the DataFrame for Profit > 0 and extract the year
t1 = DataSkincare[DataSkincare['Profit'] > 0].copy()
t1['year'] = t1['Order Date'].dt.year  # Ensure 'new_date' is in datetime format
t1 = t1[['Segment', 'Profit', 'year', 'Row ID']].sort_values(by='year')

# Step 2: Group by year and segment, and calculate max profit
t2 = (t1.groupby(['year', 'Segment'])
      .agg(profit=('Profit', 'max'))
      .reset_index())

# Step 3: Assign row numbers based on max profit within each year
t2['rn'] = t2.groupby('year')['profit'].rank(method='first', ascending=False)

# Step 4: Filter to get the rows where rn = 1
result1= t2[t2['rn'] <=3]

# Display the result
print(result1)
```

```python
result1.to_csv("profitableSegment.csv")
```

```python
#3.	What were the most profitable markets since 2020
# Step 1: Create a new column for the year
DataSkincare['year'] = DataSkincare['Order Date'].dt.year  # Ensure 'new_date' is in datetime format

# Step 2: Assign row numbers based on Profit within each year
DataSkincare['rn'] = DataSkincare.sort_values('Profit', ascending=False) \
                          .groupby('year') \
                          .cumcount() + 1  # +1 to start counting from 1

# Step 3: Filter to get the rows where rn < 6
result = DataSkincare[DataSkincare['rn'] < 6]

# Select only the relevant columns
result3 = result[['Market', 'Profit', 'year', 'rn']]

# Display the result
print(result)
```

```python
result3.to_csv("profitablemarkets.csv")
```

```python
# 4.	What were the best seller products since 2020
# Step 1: Filter the DataFrame for discount = 0 and create a new column for the year
filtered_skincare =DataSkincare[DataSkincare['Discount'] == 0].copy()
filtered_skincare['year'] = filtered_skincare['Order Date'].dt.year  # Ensure 'new_date' is in datetime format

# Step 2: Assign row numbers based on Quantity and Profit within each year
filtered_skincare['rn'] = (filtered_skincare
                           .sort_values(['Quantity', 'Profit'], ascending=[False, False])
                           .groupby('year')
                           .cumcount() + 1)  # +1 to start counting from 1

# Step 3: Filter to get the rows where rn < 2
result = filtered_skincare[filtered_skincare['rn'] < 2]

# Select only the relevant columns
result2 = result[['Product', 'Quantity', 'Profit', 'year', 'rn', 'Discount']]

print(result2)
```

```python
result2.to_csv("bestsellerproduct.csv")
```

```python
#5.	Following above, can we also view the best seller categories and subcategories?


# Step 1: Filter the DataFrame for discount = 0 and create a new column for the year
filtered_skincare = DataSkincare[DataSkincare['Discount'] == 0].copy()
filtered_skincare['year'] = filtered_skincare['Order Date'].dt.year  # Ensure 'new_date' is in datetime format

# Step 2: Assign row numbers based on Quantity and Profit within each year
filtered_skincare['rn'] = (filtered_skincare
                           .sort_values(['Quantity', 'Profit'], ascending=[False, False])
                           .groupby('year')
                           .cumcount() + 1)  # +1 to start counting from 1

# Step 3: Filter to get the rows where rn < 2
result = filtered_skincare[filtered_skincare['rn'] < 2]

# Select only the relevant columns
result4 = result[['Product', 'Quantity', 'Profit', 'year', 'rn', 'Discount', 'Category', 'Subcategory']]

# Display the result
print(result4)
```

```python
result4.to_csv("bestcategory.csv")
```

```python
# 9.	Can we list top 10 customers who spent most from each country every year?

# Assuming you have a DataFrame 'skincare' with columns: 'Customer ID', 'quantity', 'country', 'new_date', 'Sales'

# Step 1: Extract necessary columns and create 'year' column


# Step 1: Convert 'Order Date' to datetime and extract the year
DataSkincare['year'] = pd.to_datetime(DataSkincare['Order Date']).dt.year

# Step 2: Filter rows where Profit is greater than 0
DataSkincare = DataSkincare[DataSkincare['Profit'] > 0].copy()

# Step 3: Create 'rn1' with row number over partition by 'Country' and ordered by 'Sales'
DataSkincare['rn1'] = DataSkincare.sort_values(['Country', 'Sales'], ascending=[True, False])\
                                  .groupby('Country')\
                                  .cumcount() + 1

# Step 4: Filter where rn1 <= 1 to get top sales per country (t2)
t2 = DataSkincare[DataSkincare['rn1'] <= 1].copy()

# Step 5: Create 'rn2' with row number over partition by 'year' and ordered by 'Sales'
t2['rn2'] = t2.sort_values(['year', 'Sales'], ascending=[True, False])\
               .groupby('year')\
               .cumcount() + 1

# Step 6: Filter where rn2 <= 10 to get top 10 sales per year (t3)
t3 = t2[t2['rn2'] <= 10]

# Display the final result
print(t3)

```

```python
t3.to_csv("10costumers.csv")
```

```python
# 10.	Which countries contributed most profits to each category?
# Sample DataFrame (replace this with your actual DataFrame)
# DataSkincare = pd.read_csv('path_to_your_data.csv')

# Step 1: Group by Country and Category to find the maximum profit
max_profit_by_country_category = DataSkincare.groupby(['Country', 'Category'], as_index=False).agg({'Profit': 'max'})

# Step 2: Find the maximum profit for each category and the corresponding country
max_profit_per_category = max_profit_by_country_category.loc[max_profit_by_country_category.groupby('Category')['Profit'].idxmax()]

# Step 3: Rename the column for clarity
max_profit_per_category.rename(columns={'Profit': 'max_profit'}, inplace=True)

# Print the final result
print(max_profit_per_category)
```
